```python
grades=[98,91,77,83,69]

grades.append(95)
print(grades)
avg_grades=sum(grades)/len(grades)
print(f"Average grade is {avg_grades:.2f}")


highest_grade=max(grades)
lowest_grade=min(grades)
print(f"The highest grade is {highest_grade}")
print(f"The lowest grade is {lowest_grade}")

```

```python
inventory=["apples","bananas","mangos","guavas","kiwis"]

inventory.append("strawberries")

inventory.remove("kiwis")

item="strawberries"
if item in inventory:
    print(f"{item} is in stock")
else:
    print(f"{item} is not in stock")

print("Inventory list:")
for item in inventory:
    print(f"-{item}")


```
